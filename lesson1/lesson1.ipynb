### 計算直角三角形的斜邊
根據畢氏定理：$c = \sqrt{a^2 + b^2}$

```python
import math

a = 3
b = 4

# 使用公式計算斜邊 c
c = math.sqrt(a**2 + b**2)

print(f"當兩邊分別為 {a} 和 {b} 時，斜邊長度為：{c}")
```

### 計算標準差
計算數列 `[3, 6, 9, 11, 17, 25, 32]` 的標準差。

```python
import statistics

data = [3, 6, 9, 11, 17, 25, 32]
stdev = statistics.stdev(data)

print(f"數據: {data}")
print(f"標準差約為: {stdev:.2f}")
```
